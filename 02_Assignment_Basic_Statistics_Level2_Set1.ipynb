```python
#Q5
'''
Q5.	Returns on a certain business venture, to the nearest $1,000, are known to follow the following probability distribution
x	P(x)
-2,000	0.1
-1,000	0.1
0	0.2
1000	0.2
2000	0.3
3000	0.1

(i)	What is the most likely monetary outcome of the business venture?
(ii)	Is the venture likely to be successful? Explain
(iii)	What is the long-term average earning of business ventures of this kind? Explain
(iv)	What is the good measure of the risk involved in a venture of this kind? Compute this measure '''

# Given data
outcomes = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]

# (i) Most likely monetary outcome
most_likely_outcome = outcomes[probabilities.index(max(probabilities))]
print("(i) Most likely monetary outcome:", most_likely_outcome)

# (ii) Is the venture likely to be successful?
success_probability = sum(prob for outcome, prob in zip(outcomes, probabilities) if outcome > 0)
print("(ii) Probability of earning a positive return:", success_probability)

# (iii) Long-term average earnings
average_earnings = sum(outcome * prob for outcome, prob in zip(outcomes, probabilities))
print("(iii) Long-term average earnings:", average_earnings)

# (iv) Measure of risk: standard deviation
mean = sum(outcome * prob for outcome, prob in zip(outcomes, probabilities))
variance = sum(prob * (outcome - mean) ** 2 for outcome, prob in zip(outcomes, probabilities))
std_dev = variance ** 0.5
print("(iv) Measure of risk (standard deviation):", std_dev)
```
