```python
# Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. 
# We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them.
#  We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. 
# Calculate 94%,98%,96% confidence interval?


import scipy.stats as stats

# Given values
sample_mean = 200  # average weight in pounds
sample_std = 30  # standard deviation of the sample
sample_size = 2000  # sample size

# Confidence levels
confidence_levels = [0.94, 0.96, 0.98]

# Calculate critical values for each confidence level
critical_values = [stats.norm.ppf((1 + level) / 2) for level in confidence_levels]

# Calculate margin of error for each confidence level
margins_of_error = [critical_value * (sample_std / (sample_size**0.5)) for critical_value in critical_values]

# Calculate confidence intervals
confidence_intervals = [(sample_mean - margin_of_error, sample_mean + margin_of_error) for margin_of_error in margins_of_error]

# Print the results
for i, level in enumerate(confidence_levels):
    print(f"{int(level * 100)}% Confidence Interval: {confidence_intervals[i]}")
```

```python
# Q 20) Calculate probability from the given dataset for the below cases

# Data _set: Cars.csv
# Calculate the probability of MPG  of Cars for the below cases.
#        MPG <- Cars$MPG
# a.	P(MPG>38)
# b.	P(MPG<40)
# c.    P (20<MPG<50)

import pandas as pd
import numpy as np

# to read the dataset

cars = pd.read_csv(r"D:\Data_Science_course\Data Science @EXCELR\Assignments\Basic Statistics Level 1\Cars.csv")
cars_MPG = cars['MPG']
cars_MPG
mean_cars_MPG = cars_MPG.mean()
std_dev_cars_MPG = cars_MPG.std()

# z = (X  ) / 
# Finding Z score value when X = 38

Z_score_38 = (38 - mean_cars_MPG ) / std_dev_cars_MPG

# Now we have to find the probability when MPG > 38 . So it should be 1 - the probability value

probability_MPG_38 = 1- stats.norm.cdf(Z_score_38)

print(f'P(MPG>38) =  {round(probability_MPG_38,2)} or {(round(probability_MPG_38,2))*100} percent  ') 
# Finding Z score value when X = 40 

Z_score_40 = (40 - mean_cars_MPG ) / std_dev_cars_MPG
# Now we have to find the probability when MPG <  38 . So it should be  the probability value at Z score = 40 

probability_MPG_40 = stats.norm.cdf(Z_score_40)
probability_MPG_40
print(f'P(MPG<40) =  {round(probability_MPG_40,2)} or {(round(probability_MPG_40,2))*100} percent  ')
# Finding Z score value when X = 20 and X = 50 
Z_score_20 = (20 - mean_cars_MPG ) / std_dev_cars_MPG
Z_score_50 = (50 - mean_cars_MPG ) / std_dev_cars_MPG
# c.    P (20<MPG<50)
probability_20_50 = stats.norm.cdf(Z_score_50) - stats.norm.cdf(Z_score_20)
print(f'P(20<MPG<50) =  {round(probability_20_50,2)} or {(round(probability_20_50,2))*100} percent ')
```

```python
# Q 21) Check whether the data follows normal distribution
# a)	Check whether the MPG of Cars follows Normal Distribution  for Dataset: Cars.csv

import pandas as pd
import numpy as np 
# to read the dataset car
cars = pd.read_csv(r"D:\Data_Science_course\Data Science @EXCELR\Assignments\Basic Statistics Level 1\Cars.csv")

#to show the column MPG of car datset 
cars_MPG = cars['MPG']

cars_MPG

import matplotlib.pyplot as plt
import statsmodels.api as sm



fig, (ax1, ax2) = plt.subplots( 1, 2 , figsize=(12, 4))  # 1 row, 2 columns

# Histogram
ax1.hist(cars_MPG, bins=10, edgecolor='black')  # Adjust the number of bins as needed
ax1.set_title("Histogram of MPG Column")
ax1.set_xlabel("MPG Values")
ax1.set_ylabel("Frequency")

# QQ plot
sm.qqplot(cars_MPG, line='s', ax=ax2)
ax2.set_title("QQ Plot for MPG Column")
ax2.set_xlabel("Theoretical Quantiles (Standard Normal Distribution)")
ax2.set_ylabel("Sample Quantiles (MPG Column)")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()

from scipy.stats import shapiro

# Shapiro-Wilk test
stat, p_value = shapiro(cars_MPG)

# Check the p-value
if p_value > 0.05:
    print("The 'MPG' column appears to be normally distributed.")
else:
    print("The 'MPG' column does not appear to be normally distributed.")
```

```python
#   Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing.
# The sampled bulbs last an average of 260 days, with a standard deviation of 90 days.
# If the CEO's claim were true, what is the probability that 18 
# randomly selected bulbs would have an average life of no more than 260 days


from scipy.stats import t

# Given values
population_mean = 270
sample_mean = 260



sample_size = 18
sample_std_dev = 90 

# Calculate standard error of the mean (SE)
se = sample_std_dev / (sample_size**0.5)

# Calculate degrees of freedom
df = sample_size - 1

# Calculate t-Score
t_score = (sample_mean - population_mean) / se

# Find the probability using the cumulative distribution function (CDF)
probability = t.cdf(t_score, df)

# Print the result
print(f"The probability that 18 randomly selected bulbs would have an average life of no more than 260 days is {round(probability,4)} or {(round(probability,4)*100)} percent  ")
```
